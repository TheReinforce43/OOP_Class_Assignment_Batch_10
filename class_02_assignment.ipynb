Practice 04 : Bank Account 

```python
class BankAccount:

    def __init__(self, account_number,account_name, initial_balance=0):
        self.account_number = account_number
        self.balance = initial_balance
        self.account_name = account_name

    
    def deposit(self,amount):
        if amount>0:

            self.balance+=amount 
            print(f"Deposited: {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")


    def withdraw(self,amount):
        if amount>0:
            if amount <=self.balance:
                self.balance -=amount 

                print(f"Withdrew: {amount}. New balance: {self.balance}")
            else:
                print("Insufficient funds.")

        else:
            print("Withdrawal amount must be positive.")
    

    def transfer(self,request_amount,target_account):

        if request_amount<0 or self.balance < request_amount:
            print(f"Your requested amount of {request_amount} is invalid or exceeds your balance of {self.balance}.")
            return 
        

        self.withdraw(request_amount)
        target_account.deposit(request_amount)

        

    def get_balance(self):
        return self.balance



```

```python
if __name__ == "__main__":

    first_account =BankAccount("2342323","Farhan Kabir",1000)

    second_account =BankAccount("9876543","Jahidul Islam",500)

    print(f"Initial first account balance : {first_account.get_balance()}")
    print(f" Initial second account balance : {second_account.get_balance()}")

    # Deposit money
    first_account.deposit(500)
    print(f"first account balance : {first_account.get_balance()}")
    # Withdraw money
    first_account.withdraw(200)

    # transfer money 

    first_account.transfer(3000,second_account)

    print(f"Final first account balance : {first_account.get_balance()}")
    print(f"Final second account balance : {second_account.get_balance()}")


    

    
```

Practice 05 : Credit Card

```python
class CreditCard:

    max_limit = 500_000 
    day_limit = 100_000
    per_transaction_limit = 20_000

    def __init__(self):

        self.daily_spent =0 
        self.total_spent =0 


    def isCross_max_limits(self,amount):

        temp_amount = self.total_spent + amount 

        if temp_amount > self.max_limit :
            print("Transaction exceeds maximum credit limit.")
            return False
        return True
    
    def isCross_daily_limits(self,amount):

        temp_amount = self.daily_spent + amount 

        if temp_amount > self.day_limit :
            print("Transaction exceeds daily credit limit.")
            return False
        return True
    
    def isCross_per_transaction_limits(self,amount):

        if amount > self.per_transaction_limit :
            print("Transaction exceeds per transaction credit limit.")
            return False
        return True

    def make_withdrawal(self,amount):

        if amount <=0 :
            print("Withdrawal amount must be positive.")
            return 
        
        if not self.isCross_max_limits(amount):
            return 
        
        if not self.isCross_daily_limits(amount):
            return 
        
        if not self.isCross_per_transaction_limits(amount):
            return 
        
        self.daily_spent += amount 
        self.total_spent += amount 

        print(f"Withdrawal of {amount} successful. Total spent: {self.total_spent}, Daily spent: {self.daily_spent}")


    def make_bil_payment(self,amount):  

        if amount <=0 :
            print("Payment amount must be positive.")
            return 
        
        if not self.isCross_max_limits(amount):
            return 
        
      
        self.daily_spent += amount 
        self.total_spent += amount 

        print(f"Bill payment of {amount} successful. Total spent: {self.total_spent}, Daily spent: {self.daily_spent}")


```

```python
if __name__ == "__main__":

    my_card = CreditCard()

    my_card.make_withdrawal(10000)
    my_card.make_withdrawal(90000)  # This should exceed per transaction limit
    my_card.make_bil_payment(30000)
    my_card.make_bil_payment(90000)  # This should exceed daily limit
    my_card.make_withdrawal(400000)  # This should exceed max limit
```
